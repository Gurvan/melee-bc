```python
import torch.nn.functional as F
class GaussianEmbedding(nn.Module):
    def __init__(
        self,
        bounds: Tuple[float, float] = (-1, 1),
        n: int = 16,
        learnable: bool = False,
    ):
        super().__init__()
        bmin = min(bounds)
        bmax = max(bounds)
        bdif = (bmax - bmin) / (n - 3)
        self.means = torch.linspace(bmin, bmax + bdif, n, requires_grad=learnable)
        self.log_stds = (bdif * torch.ones(n, requires_grad=learnable)).log()

    def forward(self, x):
        return (
            torch.distributions.Normal(self.means, F.softplus(self.log_stds))
            .log_prob(x[..., None])
            .softmax(dim=-1)
        )

    def encode(self, x):
        return self(x)

    def decode(self, x):
        return x @ self.means
```

```python
def count_parameters(model):
    total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    suffixes = ['K', 'M', 'G', 'T', 'P']
    if total_params < 1000:
        return f"{total_params}"
    for suffix in suffixes:
        total_params /= 1000
        if total_params < 1000:
            return f"{total_params:.1f}{suffix}"
    return f"{total_params:.1f}E"
```

```python
from typing import Tuple
import torch
from torch import nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
import math
from dataset import MeleeDataloader
import models

def prepare_data(actions, states, metadata, epoch=0):
    index_self = epoch % 2
    index_other = (epoch + 1) % 2
    states1 = states[:, index_self].float()
    states2 = states[:, index_other].float()
    actions = actions[:, index_self].float()
    metadata = metadata.int()
    if index_self == 1:
        metadata = metadata[:, [0, 2, 1]]
    return (
        actions,
        states1,
        states2,
        metadata,
    )
```

```python
class BCModel(nn.Module):
    def __init__(self, *args, **kwargs) -> None:
        super().__init__(*args, **kwargs)
        state_emb_dim = 64
        action_emb_dim = 16
        latent_dim = 192
        n_head = 3
        n_layer = 2
        self.state_emb = models.StateEmbedding(state_emb_dim, latent_dim)
        self.action_emb = models.ActionEncoding(action_emb_dim)
        # self.action_proj = nn.Linear(12*emb_dim, latent_dim)
        # self.action_unproj = nn.Linear(latent_dim, 12*emb_dim)
        self.torso = nn.Transformer(
            d_model=latent_dim,
            nhead=n_head,
            num_encoder_layers=n_layer,
            num_decoder_layers=n_layer,
            activation=F.gelu,
            batch_first=True,
        )

    def loss(self, actions, obs_s, obs_o, meta):
        actions_target = actions[..., 1:,:]

        states = self.state_emb(obs_s, obs_o, meta)[..., :-1, :]
        prev_actions = self.action_emb(actions[..., :-1, :])
        # prev_actions = self.action_proj(torch.cat(prev_actions, dim=-1))
        prev_actions = torch.cat(prev_actions, dim=-1)
        logits = self.torso(states, prev_actions)
        # logits = self.action_unproj(logits).view(*prev_actions.shape[:-1], 12, -1)
        logits = logits.view(*prev_actions.shape[:-1], 12, -1)
        loss = self.action_emb.loss(logits, actions_target)
        return loss

class BCModel2(nn.Module):
    def __init__(self, *args, **kwargs) -> None:
        super().__init__(*args, **kwargs)
        state_emb_dim = 64
        action_emb_dim = 16
        latent_dim = 12 * action_emb_dim
        memory_size = 256
        self.state_emb = models.StateEmbedding(state_emb_dim, latent_dim)
        self.action_emb = models.ActionEncoding(action_emb_dim)
        # self.action_proj = nn.Linear(12*emb_dim, latent_dim)
        # self.action_unproj = nn.Linear(latent_dim, 12*emb_dim)
        self.torso = nn.GRU(latent_dim + latent_dim, memory_size, batch_first=True)
        self.action_unproj = nn.Linear(memory_size, 12*action_emb_dim)

    def loss(self, actions, obs_s, obs_o, meta):
        actions_target = actions[..., 1:,:]

        states = self.state_emb(obs_s, obs_o, meta)[..., :-1, :]
        prev_actions = self.action_emb(actions[..., :-1, :])
        # prev_actions = self.action_proj(torch.cat(prev_actions, dim=-1))
        prev_actions = torch.cat(prev_actions, dim=-1)

        logits, _ = self.torso(torch.cat((states, prev_actions), dim=-1))
        logits = self.action_unproj(logits).view(*prev_actions.shape[:-1], 12, -1)
        # logits = logits.view(*prev_actions.shape[:-1], 12, -1)
        loss = self.action_emb.loss(logits, actions_target)
        return loss

    def forward(self, actions, obs_s, obs_o, meta):
        actions_target = actions[..., 1:,:]

        states = self.state_emb(obs_s, obs_o, meta)[..., :-1, :]
        prev_actions = self.action_emb(actions[..., :-1, :])
        # prev_actions = self.action_proj(torch.cat(prev_actions, dim=-1))
        prev_actions = torch.cat(prev_actions, dim=-1)

        logits, _ = self.torso(torch.cat((states, prev_actions), dim=-1))
        logits = self.action_unproj(logits).view(*prev_actions.shape[:-1], 12, -1)
        # logits = logits.view(*prev_actions.shape[:-1], 12, -1)
        pred = self.action_emb.decode(logits)
        return pred, actions_target
```

```python
DEVICE = 'cuda'
BS = 64

model = BCModel2().to(DEVICE)

loader = MeleeDataloader(seq_len=30, path="/media/DATA/Melee/FloatData/")
data = loader.get_samples(BS, return_rewards=False, device=DEVICE)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
num_batch = loader.len(BS)
for (i, x) in enumerate(data):
    optimizer.zero_grad()
    l = model.loss(*prepare_data(*x)).mean()
    l.backward()
    optimizer.step()
    if i % 100 == 1:
        print(i, num_batch, l.item())
```
